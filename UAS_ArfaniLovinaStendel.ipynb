```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Membaca citra

```python
image = cv2.imread('buah.jpg')
```

# Konversi warna citra dari BGR ke RGB

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

# Menampilkan citra asli

```python
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.axis('off')
plt.title('Citra Asli')
plt.show()
```

## Definisikan range warna untuk deteksi buah

```python
lower_red = np.array([0, 0, 100])
upper_red = np.array([100, 100, 255])

lower_yellow = np.array([0, 100, 100])
upper_yellow = np.array([100, 255, 255])

lower_green = np.array([0, 100, 0])
upper_green = np.array([100, 255, 100])
```

## Proses deteksi buah berdasarkan warna

```python
mask_red = cv2.inRange(image, lower_red, upper_red)
mask_yellow = cv2.inRange(image, lower_yellow, upper_yellow)
mask_green = cv2.inRange(image, lower_green, upper_green)
```

### Menggabungkan semua hasil deteksi menjadi satu

```python
combined_mask = cv2.bitwise_or(mask_red, cv2.bitwise_or(mask_yellow, mask_green))
```

##### Menerapkan hasil deteksi pada citra asli dengan menerapkan colormap

```python
result_red = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_red)
result_yellow = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_yellow)
result_green = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_green)
```

### Penyesuaian kontras dan kecerahan pada hasil deteksi

```python
alpha = 1.5  # faktor kontras
beta = 30  # faktor kecerahan
adjusted_red = cv2.convertScaleAbs(result_red, alpha=alpha, beta=beta)
adjusted_yellow = cv2.convertScaleAbs(result_yellow, alpha=alpha, beta=beta)
adjusted_green = cv2.convertScaleAbs(result_green, alpha=alpha, beta=beta)
```

### Menampilkan hasil deteksi buah yang lebih jelas

```python
plt.figure(figsize=(12, 4))
```

```python
plt.subplot(141)
plt.imshow(adjusted_red)
plt.axis('off')
plt.title('Deteksi Buah Merah')
```

```python
plt.subplot(142)
plt.imshow(adjusted_yellow)
plt.axis('off')
plt.title('Deteksi Buah Kuning')
```

```python
plt.subplot(143)
plt.imshow(adjusted_green)
plt.axis('off')
plt.title('Deteksi Buah Hijau')
```

```python
plt.subplot(144)
plt.imshow(image_rgb)
plt.imshow(combined_mask, alpha=0.4)
plt.axis('off')
plt.title('Deteksi Semua Buah')
```
